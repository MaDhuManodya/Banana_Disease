```python
import os
import shutil
from bing_image_downloader import downloader
from google.colab import files

# Create a directory named 'banana' if it doesn't exist
if not os.path.exists('manzano banana'):
    os.mkdir('manzano banana')
    print("Folder 'manzano banana' created.")
else:
    print("Folder 'manzano banana' already exists.")

```

```python
# Download images of "Banana Types" into the 'banana' folder
downloader.download("manzano banana", limit=100, output_dir='manzano banana', adult_filter_off=True, force_replace=False, timeout=60)
```

```python
# Step 3: Zip the 'banana' folder
shutil.make_archive('manzano banana', 'zip', 'manzano banana')
print("Folder 'manzano banana' has been zipped as 'manzano banana.zip'.")
```
