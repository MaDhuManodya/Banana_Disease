```python
import os
import shutil
from bing_image_downloader import downloader
from google.colab import files

# Create a directory named 'banana' if it doesn't exist
if not os.path.exists('banana crown rot disease'):
    os.mkdir('banana crown rot disease')
    print("Folder 'banana crown rot disease' created.")
else:
    print("Folder 'banana crown rot disease' already exists.")

```

```python
# Download images of "Banana Types" into the 'banana' folder
downloader.download("banana crown rot disease", limit=100, output_dir='banana crown rot disease', adult_filter_off=True, force_replace=False, timeout=60)
```

```python
# Step 3: Zip the 'banana' folder
shutil.make_archive('banana crown rot disease', 'zip', 'banana crown rot disease')
print("Folder 'banana crown rot disease' has been zipped as 'banana crown rot disease.zip'.")
```
