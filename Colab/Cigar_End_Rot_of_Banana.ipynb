```python
import os
import shutil
from bing_image_downloader import downloader
from google.colab import files

# Create a directory named 'banana' if it doesn't exist
if not os.path.exists('Cigar End Rot of Banana'):
    os.mkdir('Cigar End Rot of Banana')
    print("Folder 'Cigar End Rot of Banana' created.")
else:
    print("Folder 'Cigar End Rot of Banana' already exists.")

```

```python
# Download images of "Banana Types" into the 'banana' folder
downloader.download("Cigar End Rot of Banana", limit=100, output_dir='Cigar End Rot of Banana', adult_filter_off=True, force_replace=False, timeout=60)
```

```python
# Step 3: Zip the 'banana' folder
shutil.make_archive('Cigar End Rot of Banana', 'zip', 'Cigar End Rot of Banana')
print("Folder 'Cigar End Rot of Banana' has been zipped as 'Cigar End Rot of Banana.zip'.")
```
